```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the DataFrame is already read and stored in a variable `df`
df = pd.read_csv('/content/ports data.csv')

# Convert the 'Year' column to a more usable format
df['Year'] = df['Year'].str.extract(r'(\d{4})').astype(int)

# Fill NaN values with 0 for easier plotting
df['Traffic Handled By Major Ports (UOM:t(Tonnes))'].fillna(0, inplace=True)
df['Traffic Handled By Non Major Ports (UOM:t(Tonnes))'].fillna(0, inplace=True)

# 1. Bar Plot for Traffic Handled by Major Ports by State
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Traffic Handled By Major Ports (UOM:t(Tonnes))', data=df, ci=None)
plt.xticks(rotation=90)
plt.title('Traffic Handled by Major Ports by State')
plt.xlabel('State')
plt.ylabel('Traffic Handled (Tonnes)')
plt.show()
```

```python
# 2. Bar Plot for Traffic Handled by Non-Major Ports by State
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Traffic Handled By Non Major Ports (UOM:t(Tonnes))', data=df, ci=None)
plt.xticks(rotation=90)
plt.title('Traffic Handled by Non-Major Ports by State')
plt.xlabel('State')
plt.ylabel('Traffic Handled (Tonnes)')
plt.show()
```

```python
# 3. Line Plot for Yearly Traffic Trend for Major Ports
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Traffic Handled By Major Ports (UOM:t(Tonnes))', hue='State', data=df, marker='o')
plt.title('Yearly Traffic Trend for Major Ports')
plt.xlabel('Year')
plt.ylabel('Traffic Handled (Tonnes)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 4. Line Plot for Yearly Traffic Trend for Non-Major Ports
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Traffic Handled By Non Major Ports (UOM:t(Tonnes))', hue='State', data=df, marker='o')
plt.title('Yearly Traffic Trend for Non-Major Ports')
plt.xlabel('Year')
plt.ylabel('Traffic Handled (Tonnes)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 5. Pie Chart for Traffic Distribution by State (Major Ports)
traffic_major_ports = df.groupby('State')['Traffic Handled By Major Ports (UOM:t(Tonnes))'].sum()
traffic_major_ports.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Traffic Distribution by State (Major Ports)')
plt.ylabel('')
plt.show()
```

```python
# 6. Pie Chart for Traffic Distribution by State (Non-Major Ports)
traffic_non_major_ports = df.groupby('State')['Traffic Handled By Non Major Ports (UOM:t(Tonnes))'].sum()
traffic_non_major_ports.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Traffic Distribution by State (Non-Major Ports)')
plt.ylabel('')
plt.show()
```
