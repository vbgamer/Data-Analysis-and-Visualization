```python
import numpy as np

df_np = np.array(df)

major_traffic_idx = df.columns.get_loc('Traffic Handled By Major Ports (UOM:t(Tonnes))')
non_major_traffic_idx = df.columns.get_loc('Traffic Handled By Non Major Ports (UOM:t(Tonnes))')

# Calculate the total traffic handled by major and non-major ports
total_major_traffic = np.nansum(df_np[:, major_traffic_idx].astype(float))
total_non_major_traffic = np.nansum(df_np[:, non_major_traffic_idx].astype(float))

# average traffic handled by major and non-major ports
avg_major_traffic = np.nanmean(df_np[:, major_traffic_idx].astype(float))
avg_non_major_traffic = np.nanmean(df_np[:, non_major_traffic_idx].astype(float))

# Print results
print("Total traffic handled by major ports: {total_major_traffic} tonnes")
print("Total traffic handled by non-major ports: {total_non_major_traffic} tonnes")
print("Average traffic handled by major ports: {avg_major_traffic} tonnes")
print("Average traffic handled by non-major ports: {avg_non_major_traffic} tonnes")
```

```python
import matplotlib.pyplot as plt
import numpy as np

df_np = np.array(df)

states = np.unique(df_np[:, 1])
years = np.unique(df_np[:, 2])

plt.figure(figsize=(12, 8))

for state in states:
    state_data = df_np[df_np[:, 1] == state]
    major_traffic = state_data[:, 3].astype(float)
    non_major_traffic = state_data[:, 4].astype(float)

    plt.plot(years, major_traffic, marker='o', label=f"{state} - Major Ports")
    plt.plot(years, non_major_traffic, marker='s', label=f"{state} - Non-Major Ports")

plt.xlabel('Financial Year')
plt.ylabel('Traffic (Tonnes)')
plt.title('Traffic Trends for Major and Non-Major Ports by State')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

df_np = np.array(df)

states = np.unique(df_np[:, 1])

total_major_traffic = []
total_non_major_traffic = []

for state in states:
    state_data = df_np[df_np[:, 1] == state]
    total_major_traffic.append(np.nansum(state_data[:, 3].astype(float)))
    total_non_major_traffic.append(np.nansum(state_data[:, 4].astype(float)))

plt.figure(figsize=(12, 8))
bar_width = 0.4
index = np.arange(len(states))

plt.bar(index, total_major_traffic, width=bar_width, label='Major Ports')
plt.bar(index + bar_width, total_non_major_traffic, width=bar_width, label='Non-Major Ports')

plt.xlabel('States')
plt.ylabel('Total Traffic (Tonnes)')
plt.title('Total Traffic Handled by Major and Non-Major Ports')
plt.xticks(index + bar_width/2, states, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


df_np = np.array(df)

states = ["ANDHRA PRADESH", "GOA", "TAMIL NADU"]

years = np.unique(df_np[:, 2])
state_data = {}

for state in states:
    state_data[state] = df_np[(df_np[:, 1] == state) & (~np.isnan(df_np[:, 3].astype(float))), :]

plt.figure(figsize=(12, 8))

for state in states:
    plt.plot(years, state_data[state][:, 3].astype(float), marker='o', label=state)

plt.xlabel('Financial Year')
plt.ylabel('Traffic (Tonnes)')
plt.title('Trends of Major Port Traffic Over Years')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
